```python
import pandas as pd
import numpy as np
#from tensorflow import keras
```

```python
df=pd.read_csv('PotatoPrice.csv')
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df['potato_kg'],df['price'])
plt.xlabel("Potato in Kilogram(kg)",size=15)
plt.ylabel("Pirce in Taka",size=15)
```

```python
x=df[['potato_kg']]
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# ANN Model Building : 

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(1,input_shape=[1],activation='linear')])
```

```python
model.summary()
```

# Compilation of ANN Model : 

```python
model.compile(optimizer='sgd',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

# Training of ANN Model : 

```python
model.fit(x_train,y_train,epochs=500)
```

#### There is no enough data (small dataset) to show accuracy. That's why the accuracy is not showing above.

# Prediction of ANN Model : 

```python
model.predict(x_test)
```

```python
y_test
```

# Evaluation of ANN Model (Test & Accuracy) : 

```python
test_loss,test_accuracy=model.evaluate(x_test,y_test,verbose=2)
```

### here you can clearly see the accuricy is bad, loss is high comparatively. So, if you are doing any project where  you have few data it is better to use machin learning rather artificial neural network.Here the our test data is so small that it cann't even calculate for accuracy.
